# Reading Dataset

```python
def load_perception_data(imagery_data, tsv_data, mask_filepath):
    imagery_data = imagery_data.get_data()
    selected_imagery_data = []
    labels = []
    index = 0
    
    # Reading fMRI files and TSV
    for i, row in enumerate(tsv_data.iterrows()):
        onset, duration, trial_no, event_type, img_file, response = row[1]
        if event_type == "stimulus":
            
            start_index = (onset//3)
            for k in range(start_index,  start_index+3):
                selected_imagery_data.append(imagery_data[:,:,:,k])
                        
            img_file = int(img_file.split("_")[0].replace("n", ""))
            for j in range(index, index+3):
                labels.append(img_file)
                index += 1
    
    selected_imagery_data = np.array(selected_imagery_data)
    labels = np.array(labels)

    # Making a common mask
    mask_files = os.listdir(mask_filepath)
    aggregate_features = np.zeros((165,1))
    feature_count = 0
    all_roi_regions = np.zeros((64,64,50))

    for filename in mask_files:
        mask_filename = mask_filepath + filename
        title = '_'.join(filename.split(".", 1)[0].split("_")[2:4])
        
        if("V1" in title or "V2" in title or "V3" in title or "V4" in title or "HVC" in title):
            mask_file = nib.load(mask_filename).get_data()
            all_roi_regions += mask_file
    
    all_roi_regions[all_roi_regions!=0] = 1
    
    # Applying the mask and concatenating data
    temp_list = []
    for i in range(selected_imagery_data.shape[0]):
        temp = selected_imagery_data[i,:,:,:] + 0.01
        roi_data = temp * all_roi_regions
        roi_data = roi_data[roi_data != 0]
        roi_data = roi_data - 0.01
        temp_list.append(roi_data)

    temp_list = np.array(temp_list)
    
    return temp_list, labels
```

```python
def load_imagery_data(imagery_data, tsv_data, mask_filepath):
    imagery_data = imagery_data.get_data()    
    selected_imagery_data = []
    labels = []
    index = 0
    
    # Reading fMRI data and TSV
    for i, row in enumerate(tsv_data.iterrows()):
        onset, duration, trial_no, event_type, img_file, response, evaluation = row[1]
        if event_type == "imagery":
            
            start_index = (onset//3)
            for k in range(start_index,  start_index+5):
                selected_imagery_data.append(imagery_data[:,:,:,k])
                        
            img_file = int(img_file.replace("n", ""))
            for j in range(index, index+5):
                labels.append(img_file)
                index += 1
    
    selected_imagery_data = np.array(selected_imagery_data)
    labels = np.array(labels)
    
    # Making a common mask 
    mask_files = os.listdir(mask_filepath)
    aggregate_features = np.zeros((25,1))
    feature_count = 0
    all_roi_regions = np.zeros((64,64,50))
    
    for filename in mask_files:
        mask_filename = mask_filepath + filename
        title = '_'.join(filename.split(".", 1)[0].split("_")[2:4])
        if("V1" in title or "V2" in title or "V3" in title or "V4" in title or "HVC" in title):
            mask_file = nib.load(mask_filename).get_data()
            all_roi_regions += mask_file
            
    all_roi_regions[all_roi_regions!=0] = 1
    
    # Applying mask to the data
    temp_list = []
    for i in range(selected_imagery_data.shape[0]):
        temp = selected_imagery_data[i,:,:,:] + 0.01
        roi_data = temp * all_roi_regions
        roi_data = roi_data[roi_data != 0]
        roi_data = roi_data - 0.01
        temp_list.append(roi_data)

    temp_list = np.array(temp_list)
    
    return temp_list, labels
```

# Imagery data

```python
X_imagery = []
y_imagery = []
subject_data_path = "Dataset/sub-01/"
subject_data_files = os.listdir(subject_data_path)

for root in subject_data_files:
    if("ses-imageryTest" in root):
        imagery_data_path = subject_data_path + root + "/func/"
        imagery_data_files = os.listdir(imagery_data_path)
        
        imager_data_file_names = []
        for data_file in imagery_data_files:
            file_name ='_'.join(data_file.split("_")[0:-1])
            imager_data_file_names.append(file_name)

        imager_data_file_names = sorted(set(imager_data_file_names))

        for file_name in imager_data_file_names:
            fMRI_file_name = file_name + "_bold.nii.gz"
            tsv_file_name = file_name + "_events.tsv"
            mask_filepath = "Subject-01/Anat/"

            imagery_data = nib.load(imagery_data_path + fMRI_file_name)

            df = pd.read_csv(imagery_data_path + tsv_file_name, delimiter="\t")
            imagery_data, labels = load_imagery_data(imagery_data, df, mask_filepath)

            X_imagery.append(imagery_data)
            y_imagery.append(labels)


            print(f"File name: {file_name}, imagery_data shape: {imagery_data.shape}, labels shape: {labels.shape}")
    
X_imagery = np.array(X_imagery)
X_imagery = X_imagery.reshape((X_imagery.shape[0]*X_imagery.shape[1], X_imagery.shape[2]))

y_imagery = np.array(y_imagery)
y_imagery = y_imagery.reshape((y_imagery.shape[0]*y_imagery.shape[1],))

print(f"X_imagery shape: {X_imagery.shape}, y_imagery shape: {y_imagery.shape}")
    
```

# Perception Data

```python
X_percept = []
y_percept = []

subject_data_path = "Dataset/sub-01/"
subject_data_files = os.listdir(subject_data_path)

for root in subject_data_files:
    if("ses-perceptionTest" in root):
        perception_data_path = subject_data_path + root + "/func/"
        perception_data_files = os.listdir(perception_data_path)

        perception_data_file_names = []
        for data_file in perception_data_files:
            file_name ='_'.join(data_file.split("_")[0:-1])
            perception_data_file_names.append(file_name)

        perception_data_file_names = sorted(set(perception_data_file_names))

        for file_name in perception_data_file_names:
            fMRI_file_name = file_name + "_bold.nii.gz"
            tsv_file_name = file_name + "_events.tsv"
            mask_filepath = "Subject-01/Anat/"

            perception_data = nib.load(perception_data_path + fMRI_file_name)

            df = pd.read_csv(perception_data_path + tsv_file_name, delimiter="\t")
            perception_data, labels = load_perception_data(perception_data, df, mask_filepath)

            X_percept.append(perception_data)
            y_percept.append(labels)


            print(f"File name: {file_name}, perception_data shape: {perception_data.shape}, labels shape: {labels.shape}")
    
X_percept = np.array(X_percept)
X_percept = X_percept.reshape((X_percept.shape[0]*X_percept.shape[1], X_percept.shape[2]))

y_percept = np.array(y_percept)
y_percept = y_percept.reshape((y_percept.shape[0]*y_percept.shape[1],))

print(f"X_percept shape: {X_percept.shape}, y_percept shape: {y_percept.shape}")
    
```

```python
def preProcessData(X, y, mode):
    scaler = StandardScaler()
    encoder = LabelEncoder()
    if (mode == "Combined"):
        scaler.fit(X)
        X = scaler.transform(X)
        # encode class values as integers
        encoder.fit(y)
        y = encoder.transform(y)
        
        print(f"[{mode}] X: {X.shape}, y: {y.shape}")
        print(f"Number of classes in y: {len(np.unique(y))}")
        
        return X, y
    
    else:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

        scaler.fit(X_train)
        X_train = scaler.transform(X_train)
        X_test = scaler.transform(X_test)

        # encode class values as integers

        encoder.fit(y_train)
        y_train = encoder.transform(y_train)
        y_test = encoder.transform(y_test)

        print(f"[{mode}] Train_X: {X_train.shape}, Train_y: {y_train.shape}, X_test: {X_test.shape}, y_test: {y_test.shape}")
        print(f"Number of classes in Train: {len(np.unique(y_train))} , Number of classes in Test: {len(np.unique(y_test))}")

        return X_train, X_test, y_train, y_test
```

### Get Perception (pt) & Imagery (im) Data

```python
X_train_pt, X_test_pt, y_train_pt, y_test_pt = preProcessData(X_percept, y_percept, "Perception")
X_train_im, X_test_im, y_train_im, y_test_im = preProcessData(X_imagery, y_imagery, "Imagery")

# combine data
combined_X = np.concatenate((X_train_pt, X_train_im), axis=0)
combined_y = np.hstack((y_train_pt, y_train_im))

combined_X, combined_y = shuffle(combined_X, combined_y) # since they are concatenated at the last, shuffle and preserve indices
X_train_cb, y_train_cb = preProcessData(combined_X, combined_y, "Combined")
```

## AlexNet

```python
def alexNet(pretrained_weights = None,input_size = (4466,)):
    
    inputs = Input(input_size)
    dense1 = Dense(64, activation='relu', kernel_initializer = 'he_normal')(inputs)
    bn1 = BatchNormalization()(dense1)
    dropout1 = Dropout(0.5)(bn1)
    dense2 = Dense(128, activation='relu', kernel_initializer = 'he_normal')(dropout1)
    bn2 = BatchNormalization()(dense2)
    dropout2 = Dropout(0.5)(bn2)
    dense3 = Dense(512, activation='relu',kernel_initializer = 'he_normal')(dropout2)
    bn3 = BatchNormalization()(dense3)
    dropout3 = Dropout(0.5)(bn3)
    dense4 = Dense(512, activation='relu',kernel_initializer = 'he_normal')(dropout3)
    dense5 = Dense(256, activation='relu',kernel_initializer = 'he_normal')(dense4)
    dropout4 = Dropout(0.5)(dense5)
    dense6 = Dense(128, activation='relu',kernel_initializer = 'he_normal')(dropout4)
    dense7 = Dense(64, activation='relu',kernel_initializer = 'he_normal')(dense6)
    bn7 = BatchNormalization()(dense7)

    dense8 = Dense(50, activation='softmax')(bn7)
    model = Model(input = inputs, output = dense8)
    model.compile(loss='sparse_categorical_crossentropy', optimizer = keras.optimizers.Adam(lr = 1e-4), metrics=['accuracy'])
    
    if(pretrained_weights):
        print('loading_weights')
        model.load_weights(pretrained_weights)

    return model
```

```python
def fit_alexNet(X_train, X_test, y_train, y_test):
    model = alexNet(input_size=(X_train.shape[1],))
    print(model.summary())
    model = alexNet(input_size=(X_train.shape[1],))
    history = model.fit(X_train, y_train, batch_size=64, epochs=500, validation_data=(X_test, y_test), shuffle=True, verbose=2)
    
    score = model.evaluate(X_test, y_test, batch_size=16)    
    return score, history

def plot_acc(history, file_name):
    plt.plot(history.history['acc'], label="train")
    plt.plot(history.history['val_acc'], label="val")
    plt.title("Accuracy Curve Train vs Val")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.savefig(file_name, dpi=300)
    plt.show()
    
def plot_loss(history, file_name):
    plt.plot(history.history['loss'], label="train")
    plt.plot(history.history['val_loss'], label="val")
    plt.title("Loss Curve Train vs Test")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.savefig(file_name, dpi=300)
    plt.show()
```

## 1-Layer MLP

```python
def one_layer_mlp(input_size = (4466,)):
    inputs = Input(input_size)
    dense1 = Dense(64, activation='relu', kernel_initializer = 'he_normal')(inputs)
    bn1 = BatchNormalization()(dense1)
    dropout1 = Dropout(0.5)(bn1)
    dense2 = Dense(50, activation='softmax')(dropout1)

    model = Model(input = inputs, output = dense2)
    model.compile(loss='sparse_categorical_crossentropy', optimizer = keras.optimizers.Adam(lr = 1e-4), metrics=['accuracy'])

    return model
```

```python
def fit_one_layer_mlp(X_train, X_test, y_train, y_test):
    model = one_layer_mlp(input_size=(X_train.shape[1],))
    model.summary()
    model = one_layer_mlp(input_size=(X_train.shape[1],))
    history = model.fit(X_train, y_train, batch_size=16, epochs=100, validation_data=(X_test, y_test), shuffle=True, verbose=2)
    
    score = model.evaluate(X_test, y_test, batch_size=16)    
    return score, history

def plot_acc(history, file_name):
    plt.plot(history.history['acc'], label="train")
    plt.plot(history.history['val_acc'], label="val")
    plt.title("Accuracy Curve Train vs Val")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.savefig(file_name, dpi=300)
    plt.show()
    
def plot_loss(history, file_name):
    plt.plot(history.history['loss'], label="train")
    plt.plot(history.history['val_loss'], label="val")
    plt.title("Loss Curve Train vs Test")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.savefig(file_name, dpi=300)
    plt.show()
```
