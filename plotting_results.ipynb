```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.color_palette("pastel")

%matplotlib inline
```

```python
# pre-processed fMRI data

subjs = ["Subject-01", "Subject-02", "Subject-03"]
log_reg_pt_pt = [96, 75.6, 94.2]
fc_net_pt_pt = [54.6, 36.15, 67.13]
mlp_pt_pt = [71.7, 58.1, 85.6]

log_reg_im_im = [0.0, 1.2, 3.6]
fc_net_im_im = [0.0, 1.2, 1.2]
mlp_im_im = [1.2, 1.2, 4.2]

log_reg_pt_im = [2.4, 1.8, 6.0]
fc_net_pt_im = [1.2, 4.8, 6.0]
mlp_pt_im = [1.8, 4.2, 4.8]

log_reg_pt_im_pt = [91.0, 64.1, 90.3]
fc_net_pt_im_pt = [55.1, 32.6, 60.2]
mlp_pt_im_pt = [63.8, 48.7, 79.7]

log_reg_pt_im_im = [0.0, 1.2, 4.2]
fc_net_pt_im_im = [2.4, 3.6, 4.2]
mlp_pt_im_im = [1.2, 3.0, 4.2]

x = np.arange(3)
```

```python
width = 0.2
ax = plt.subplot(111)
ax.bar(x, log_reg_pt_pt, width=0.2, color='tab:cyan', align='center', label="Logistic Regression")
ax.bar(x+width, fc_net_pt_pt, width=0.2, color='tab:orange', align='center', label="8-Layer FC")
ax.bar(x + (2 * width), mlp_pt_pt, width=0.2, color='tab:red', align='center', label="2-layer MLP")
ax.set_xticks(x + width + width/2)
ax.set_xticklabels(subjs)
plt.title("Pre-processed Data Trained on Pt., Tested on Pt.")
plt.xlabel("Subjects")
plt.ylabel("Test Accuracy (percent)")
sns.set(rc={'text.usetex': True})
plt.legend()
plt.savefig("pre_train_pt_test_pt.png", dpi=300)
plt.show()
```
